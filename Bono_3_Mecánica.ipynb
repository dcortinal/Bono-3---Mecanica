```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega = 1/4
r0 = 1.0
t = np.linspace(0, 100, int(1e5))

r = (r0 / 2) * (np.exp(omega * t) + np.exp(-omega * t))
phi = omega * t

x = r*np.cos(phi)
y = r*np.sin(phi)
```

```python
ffig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Gr치fico [1,1]
axs[0, 0].plot(t, r,c='k')
axs[0, 0].set_title("r(t) vs t", fontsize=20)
axs[0, 0].set_xlabel('t [u.t.]', fontsize=16)
axs[0, 0].set_ylabel('r(t) [u.r.]', fontsize=16)
axs[0, 0].grid(True)

# Gr치fico [1,2]
axs[0, 1].plot(t, phi,c='k')
axs[0, 1].set_title("$\\phi$(t) vs t", fontsize=20)
axs[0, 1].set_xlabel('t [u.t.]', fontsize=16)
axs[0, 1].set_ylabel('$\\phi$(t) [u.a.]', fontsize=16)
axs[0, 1].grid(True)

# Gr치fico [2,1]
axs[1, 0].plot(x, y,c='k')
axs[1, 0].set_title("Espiral en Cartesianas", fontsize=20)
axs[1, 0].set_xlabel('$x(r,\\phi)$', fontsize=16)
axs[1, 0].set_ylabel('$y(r,\\phi)$', fontsize=16)
axs[1, 0].set_xscale("symlog")
axs[1, 0].set_yscale("symlog")
axs[1, 0].grid(True)

# Gr치fico [2,2]
axs[1, 1].plot(phi, r,c='k')
axs[1, 1].set_title("r(t) vs $\\phi$(t)", fontsize=20)
axs[1, 1].set_xlabel('$\\phi$(t) [u.a.]', fontsize=16)
axs[1, 1].set_ylabel('r(t) [u.r.]', fontsize=16)
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
